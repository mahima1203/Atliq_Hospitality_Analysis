# 11.Occupancy % by week number

```python
# 11.Occupancy % by week number
import matplotlib.pyplot as plt

# Group data and calculate mean occupancy percentage
data = df_merged.groupby("week no")["occ%"].mean().round(2)

# Plot the line chart
plt.figure(figsize=(8,4))
plt.plot(data.index, data.values, marker="o", linestyle="-", color="b", label="Occupancy %")

# Add data labels on each point
for x, y in zip(data.index, data.values):
    plt.text(x, y, f"{y}%", fontsize=10, ha='center', va='bottom', color="black")

# Labels and title
plt.title("Occupancy % by Week No")
plt.xlabel("Week No")
plt.ylabel("Occupancy %")
plt.xticks(rotation=45)  # Rotate x labels if needed
plt.grid(False)
plt.legend()
plt.show()
```
